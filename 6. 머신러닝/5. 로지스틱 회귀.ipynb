# **시그모이드 함수(Sigmoid function)**  
화면 캡처 2022-05-24 143835.png  

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 시그모이드 함수를 그래프로 시각화
def sigmoid(x):
  return 1 / (1 + np.exp(-x))
```

```python
x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)
```

```python
plt.plot(x, y, 'g')
plt.plot([0, 0], [1.0, 0.0], ':') # 가운데 점선 추가
plt.title("Sigmoid Function")
plt.show()
```

- 시그모이드 함수는 출력값을 0과 1사이의 값으로 조정한다.

```python
# 가중치를 변화시켰을 때의 그래프 모양
x = np.arange(-5.0, 5.0, 0.1)
y1 = sigmoid(0.5*x)
y2 = sigmoid(x)
y3 = sigmoid(2*x)

plt.plot(x, y1, 'r', linestyle='--') # w의 값이 0.5일때
plt.plot(x, y2, 'g') # w의 값이 1일때
plt.plot(x, y3, 'b', linestyle='--') # w의 값이 2일때
plt.plot([0,0],[1.0,0.0], ':') # 가운데 점선 추가
plt.title('Sigmoid Function')
plt.show()
```

- w의 값이 커지면 경사가 커지고 w의 값이 작아지면 경사가 작아진다.

```python
# 편향 b의 값에 따라서 그래프 변화
x = np.arange(-5.0, 5.0, 0.1)
y1 = sigmoid(x+0.5)
y2 = sigmoid(x+1)
y3 = sigmoid(x+1.5)

plt.plot(x, y1, 'r', linestyle='--') # x + 0.5
plt.plot(x, y2, 'g') # x + 1
plt.plot(x, y3, 'b', linestyle='--') # x + 1.5
plt.plot([0,0],[1.0,0.0], ':') # 가운데 점선 추가
plt.title('Sigmoid Function')
plt.show()
```

- b 값에 따라서 그래프가 이동한다.
