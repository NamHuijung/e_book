# **머신 러닝 모델의 평가**  
머신 러닝을 위한 데이터를 준비했다면 기계를 학습하기 전 해당 데이터를 훈련용, 검증용, 테스트용 이렇게 세 가지로 분리하는 것이 일반적이다.  
- 훈련 데이터 : 머신 러닝 모델을 학습하는 용도
- 검증 데이터 : 모델의 성능을 조정하기 위한 용도  
모델이 훈련 데이터에 과적합(overfitting)이 되고 있는지 판단하거나 하이퍼파라미터의 조정을 위한 용도
- 테스트 데이터 : 모델의 성능을 평가하기 위한 용도 

하이퍼파라미터(초매개변수) : 모델의 성능에 영향을 주는 사람이 값을 지정하는 변수, 예) 학습률, 뉴런의 수, 층의 수 등  
매개변수 : 학습을 하는 동안 값이 계속해서 변하는 수, 예) 가중치와 편향 등

# **분류(Classification)와 회귀(Regression)**

### **분류**
**1. 이진 분류 문제(Binary Classification)**  
주어진 입력에 대해서 두 개의 선택지 중 하나의 답을 선택해야 하는 경우  
**2. 다중 클래스 분류(Multi-class Classification)**  
주어진 입력에 대해서 세 개 이상의 선택지 중에서 답을 선택해야 하는 경우  

### **회귀**  
어떠한 연속적인 값의 범위 내에서 예측값이 나오는 경우  
회귀 문제의 예시로 시계열 데이터(Time Series Data)를 이용한 주가 예측, 생산량 예측, 지수 예측 등이 있다.

# **지도 학습과 비지도 학습**
**1. 지도 학습(Supervised Learning)**
레이블(Label)이라는 정답과 함께 학습하는 것  
자연어 처리는 대부분 지도 학습에 속한다.  
**2. 비지도 학습(Unsupervised Learning)**
데이터에 별도의 레이블이 없이 학습하는 것  
예를 들어 텍스트 처리 분야의 토픽 모델링 알고리즘인 LSA나 LSD는 비지도 학습에 속한다.  
**3. 자기지도 학습(Self-Supervised Learning, SSL)**
레이블이 없는 데이터가 주어지면, 모델이 학습을 위해서 스스로 데이터로부터 레이블을 만들어서 학습하는 경우  
대표적인 예시로는 Word2Vec와 같은 워드 임베딩 알고리즘이나 BERT와 같은 언어 모델의 학습 방법을 들 수 있다.

# **샘플(Sample)과 특성(Feature)**  
머신러닝에서 하나의 데이터를 행렬 관점에서는 하나의 행을 샘플(Sample)이라고 부른다.  
그리고 종속 변수 y를 예측하기 위한 각각의 독립 변수 x를 특성(Feature)이라고 부른다. 행렬 관점에서 각 열에 해당된다.

# **혼동 행렬(Confusion Matrix)**
- True Positive(TP) : 실제 True인 정답을 True라고 예측 (정답)
- False Positive(FP) : 실제 False인 정답을 True라고 예측 (오답)
- False Negative(FN) : 실제 True인 정답을 False라고 예측 (오답)
- True Negative(TN) : 실제 False인 정답을 False라고 예측 (정답)

# **과적합(Overfitting)과 과소 적합(Underfitting)**
- 과적합 : 훈련 데이터를 과하게 학습한 경우.  
과적합 상황에서는 훈련 데이터에 대해서는 오차가 낮지만, 테스트 데이터에 대해서는 오차가 커진다.  
- 과소 적합 : 테스트 데이터의 성능이 올라갈 여지가 있음에도 훈련을 덜 한 상태  
과소 적합은 훈련 자체가 부족한 상태이므로 훈련 횟수인 에포크가 지나치게 적으면 발생할 수 있다.

### **과적합 방지를 고려한 일반적인 딥러닝 모델의 학습 과정**  
1. 주어진 데이터를 훈련 데이터, 검증 데이터, 테스트 데이터로 나눈다.(6 : 2 : 2)  
2. 훈련 데이터로 모델을 학습한다.
3. 검증 데이터로 모델을 평가하여 검증 데이터에 대한 정확도와 오차(loss)를 계산한다.  
4. 검증 데이터의 오차가 증가하였다면 과적합 징후이므로 학습 종료 후 다음 단계 또는 2단계로 재이동한다.
5. 모델의 학습이 종료되었으니 테스트 데이터로 모델을 평가한다.
