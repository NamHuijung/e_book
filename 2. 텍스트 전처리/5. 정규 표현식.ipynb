# **정규 표현식 실습**

```python
import re
```

```python
# . 기호
# 한 개의 임의의 문자를 나타낸다.

r = re.compile("a.c") # a와 c 사이에는 어떤 1개의 문자라고 올 수 있다. akc, azc, avc,...
print(r.search("kkk")) # 아무런 결과도 출력되지 않는다.
print(r.search("abc"))
```

```python
# ? 기호
# ?앞의 문자가 존재할 수도 있고 존재하지 않을 수도 있는 경우를 나타낸다.

r = re.compile("ab?c") # b는 있다고 취급할 수도 있고 없다고 취급할 수도 있다.
print(r.search("abbc")) # 아무런 결과도 출력되지 않는다.
print(r.search("abc"))
print(r.search("ac"))
```

```python
# * 기호
# 바로 앞의 문자가 0개 이상일 경우를 나타낸다.

r = re.compile("ab*c") # b의 개수는 무수히 많을 수 있다.
print(r.search("a")) # 아무런 결과도 출력되지 않는다.
print(r.search("ac"))
print(r.search("abc"))
print(r.search("abbbbc"))
```

```python
# + 기호
# *와 유사하지만 다른점은 앞의 문자가 최소 1개 이상이어야 한다.

r = re.compile("ab+c")
print(r.search("ac")) # 아무런 결과도 출력되지 않는다.
print(r.search("abc"))
print(r.search("abbbbc"))
```

```python
# ^ 기호
# 시작되는 문자열을 지정한다.

r = re.compile("^ab") # 문자열 ab로 시작되는 경우 매치

print(r.search("bbc")) # 아무런 결과도 출력되지 않는다.
print(r.search("zab")) # 아무런 결과도 출력되지 않는다.
print(r.search("abz"))
```

```python
# {숫자} 기호
# 문자에 해당 기호를 붙이면, 해당 문자를 숫자만큼 반복한 것을 나타낸다.

r = re.compile("ab{2}c") # a와 c 사이에 b가 존재하면서 b가 2개인 문장열에 대해서 매치

print(r.search("ac")) # 아무런 결과도 출력되지 않는다.
print(r.search("abc")) # 아무런 결과도 출력되지 않는다.
print(r.search("abbbbbc")) # 아무런 결과도 출력되지 않는다.
print(r.search("abbc"))
```

```python
# {숫자1, 숫자2} 기호
# 문자에 해당 기호를 붙이면, 해당 문자를 숫자1 이상 숫자 2 이하만큼 반복한다.

r = re.compile("ab{2,8}c") # a와 c 사이에 b가 존재하면서 b는 2개 이상 8개 이하인 문자열에 대해서 매치

print(r.search("ac")) # 아무런 결과도 출력되지 않는다.
print(r.search("abc")) # 아무런 결과도 출력되지 않는다.
print(r.search("abbbbbbbbbc")) # 아무런 결과도 출력되지 않는다.
print(r.search("abbc"))
print(r.search("abbbbbbbbc"))
```

```python
# {숫자, } 기호
# 문자에 해당 기호를 붙이면 해당 문자를 숫자 이상 만큼 반복한다.

r = re.compile("a{2,}bc") # 뒤에 bc가 붙으면서 a의 개수가 2개 이상인 경우인 문자열과 매치 

print(r.search("bc")) # 아무런 결과도 출력되지 않는다.
print(r.search("aa")) # 아무런 결과도 출력되지 않는다.
print(r.search("aabc"))
print(r.search("aaaaaaaabc"))
```

```python
# [] 기호
# []안에 문자들을 넣으면 그 문자들 중 한 개의 문자와 매치라는 의미를 가진다.

r = re.compile("[abc]") # [abc]는 [a-c]와 같다. a 또는 b 또는 c가 들어가있는 문자열과 매치
print(r.search("zzz")) # 아무런 결과도 출력되지 않는다.
print(r.search("a"))
print(r.search("aaaaaaa"))
print(r.search("baac"))
```

```python
r = re.compile("[a-z]") # 알파벳 소문자에 대해서 범위 지정

print(r.search("AAA")) # 아무런 결과도 출력되지 않는다.
print(r.search("111")) # 아무런 결과도 출력되지 않는다.
print(r.search("aBC"))
```

```python
# [^문자] 기호
# ^ 기호 뒤에 붙은 문자들을 제외한 모든 문자를 매치하는 역할을 한다.

r = re.compile("[^abc]") # a 또는 b 또는 c가 들어간 문자열을 제외한 모든 문자열ㅇ르 매치

print(r.search("a")) # 아무런 결과도 출력되지 않는다.
print(r.search("ab")) # 아무런 결과도 출력되지 않는다.
print(r.search("b")) # 아무런 결과도 출력되지 않는다.
print(r.search("d"))
print(r.search("1")  )
```

# **정규 표현식 모듈 함수 실습**

**re.match()와 re.search()의 차이**  
search()가 정규 표현식 전체에 대해서 문장열이 매치하는지를 본다면, match()는 문자열의 첫 부분부터 정규 표현식과 매치하는지를 확인한다.  
문자열 중간에 찾을 패턴이 있더라도 match 함수는 문자열의 시작에서 패턴이 일치하지 않으면 찾지 않는다.

```python
# re.compile()와 re.match()

r = re.compile("ab.")

print(r.match("kkkabc")) # 아무런 결과도 출력되지 않는다.

print(r.search("kkkabc")  )
print(r.match("abckkk"))
```

```python
# re.split()
# 입력된 정규 표현식을 기준으로 문자열들을 분리하여 리스트로 리턴한다.

# 공백 기준 분리
text = "사과 딸기 수박 메론 바나나"
re.split(" ", text)
```

```python
# 줄바꿈 기준 분리
text = """사과
딸기
수박
메론
바나나"""

re.split("\n", text)
```

```python
# '+'를 기준으로 분리
text = "사과+딸기+수박+메론+바나나"

re.split("\+", text)
```

```python
# re.findall()
# 정규표현식과 매치되는 모든 문자열들을 리스트로 리턴한다.
# 단, 매치되는 문자열이 없다면 빈 리스트를 리턴한다.

text = """이름 : 김철수
전화번호 : 010 - 1234 - 1234
나이 : 30
성별 : 남"""

re.findall("\d+", text)
```

```python
# 입력 텍스트에 숫자 X
re.findall("\d+", "문자열입니다.")
```

```python
# re.sub()
# 정규 표현식 패턴과 일치하는 문자열을 찾아 다른 문자열로 대체할 수 있다.

# 알파벳 외의 문자는 공백으로 처리
text = "Regular expression : A regular expression, regex or regexp[1] (sometimes called a rational expression)[2][3] is, in theoretical computer science and formal language theory, a sequence of characters that define a search pattern."

preprocessed_text = re.sub('[^a-zA-Z]', ' ', text)
print(preprocessed_text)
```

# **정규 표현식 텍스트 전처리 예제**

```python
text = """100 John    PROF
101 James   STUD
102 Mac   STUD"""
```

```python
# \s+는 공백을 찾아내는 정규 표현식
# 뒤에 붙는 +는 최소 1개 이상의 패턴을 찾아낸다는 의미이다.
# s는 공백을 의미하기 때문에 최소 1개 이상의 공백인 패턴을 찾아낸다.
re.split('\s+', text)
```

```python
# 숫자만 뽑아오기
# \d는 숫자에 해당되는 정규표현식
re.findall('\d+',text) # 일치하는 값을 찾아낸다.
```

```python
# 대문자인 행의 값만 가져오기
re.findall('[A-Z]',text)
```

```python
# 대문자가 연속으로 4번 등장하는 경우
re.findall('[A-Z]{4}',text)
```

```python
# 대문자로 구성된 문자열을 가져온다.
# 처음에 대문자가 등장한 후 소문자가 여러번 등장하는 경우에 매치
re.findall('[A-Z][a-z]+',text)
```

# **정규 표현식을 이용한 토큰화**

```python
from nltk.tokenize import RegexpTokenizer

text = "Don't be fooled by the dark sounding name, Mr. Jone's Orphanage is as cheery as cheery goes for a pastry shop"

# RegexpTokenizer()에서 괄호 안에 하나의 토큰으로 규정하기를 원하는 정규 표현식을 넣어서 토큰화를 수행
tokenizer1 = RegexpTokenizer("[\w]+") # \w+ : 문자 또는 숫자가 1개 이상인 경우 의미
tokenizer2 = RegexpTokenizer("\s+", gaps=True) # 공백을 기준으로 토큰화
# gaps = True : 해당 정규 표현식을 토큰으로 나누기 위한 기준으로 사용한다는 의미이다.
# 만약 기재하지 않는다면 토큰화의 결과는 공백들만 나오게 된다.

print(tokenizer1.tokenize(text))
print(tokenizer2.tokenize(text)) # 아포스트로피나 온점을 제외하지 않고 토큰화가 수행됨
```
