```python
import numpy as np
```

### **np.array()**  
리스트, 튜플, 배열로부터 ndarray를 생성한다.

```python
# 1차원 배열
vec = np.array([1, 2, 3, 4, 5])
print(vec)

print("vec의 축의 개수 :", vec.ndim) # 축의 개수 출력
print("vec의 크기(shape) :", vec.shape) # 크기 출력
```

```python
# 2차원 배열
mat = np.array([[10, 20, 30], [60, 70, 80]])
print(mat)

print("mat의 축의 개수 :", mat.ndim) # 축의 개수 출력
print("mat의 크기(shape) :", mat.shape) # 크기 출력
```

```python
# 타입 확인
print("vec의 타입 :", type(vec))
print("mat의 타입 :", type(mat))
```

### **ndarray의 초기화**

```python
# 모든 값이 0인 2X3 배열 생성
zero_mat = np.zeros((2, 3))
print(zero_mat)
```

```python
# 모든 값이 1인 2X3 배열 생성
one_mat = np.ones((2, 3))
print(one_mat)
```

```python
# 모든 값이 특정 상수인 배열 생성
same_value_mat = np.full((2, 2), 2)
print(same_value_mat)
```

```python
# 대각선 값이 1이고 나머지 값이 0인 2차원 배열 생성
eye_mat = np.eye(3)
print(eye_mat)
```

```python
# 임의의 값으로 채워진 배열 생성
random_mat = np.random.random((2, 3))
print(random_mat)
```

### **np.arange()**  
0부터 (n-1)까지의 값을 가지는 배열을 생성

```python
# 0부터 9까지
range_vec = np.arange(10)
print(range_vec)
```

```python
# 1부터 9까지 +2씩 적용되는 범위
n = 2
range_n_step_vec = np.arange(1, 10, n)
print(range_n_step_vec)
```

### **np.reshape()**  
내부 데이터는 변경하지 않으면서 배열의 구조를 바꾼다.

```python
reshape_mat = np.array(np.arange(30)).reshape((5, 6))
print(reshape_mat)
```

### **Numpy 슬라이싱**

```python
mat = np.array([[1, 2, 3], [4, 5, 6]])
print(mat)
```

```python
# 첫번째 행 출력
slicing_mat = mat[0, :]
print(slicing_mat)
```

```python
# 두번째 열 출력
slicing_mat = mat[:, 1]
print(slicing_mat)
```

### **Numpy 정수 인덱싱**

```python
mat = np.array([[1, 2], [4, 5], [7, 8]])
print(mat)
```

```python
# 1행 0열의 원소
# 0부터 카운트하므로 두번째 행 첫번째 열의 원소
print(mat[1, 0])
```

```python
# mat[[2행, 1행], [0열, 1열]]
indexing_mat = mat[[2, 1], [0, 1]]
print(indexing_mat)
```

### **Numpy 연산**

```python
x = np.array([1, 2, 3])
y = np.array([4, 5, 6])
```

```python
# result = np.add(x, y)
result = x + y
print(result)
```

```python
# result = np.subtract(x, y)
result = x - y
print(result)
```

```python
# result = np.multiply(result, x)
result = result * x
print(result)
```

```python
# result = np.divide(result, x)
result = result / x
print(result)
```

```python
# 행렬 곱
mat1 = np.array([[1, 2], [3, 4]])
mat2 = np.array([[5, 6], [7, 8]])
mat3 = np.dot(mat1, mat2)
print(mat3)
```
